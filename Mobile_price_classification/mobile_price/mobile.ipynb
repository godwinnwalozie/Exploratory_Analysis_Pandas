# Lets just play around performing kmeans on the dataset to find our own optimal clusters 

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
```

```python
X_scaled = sc.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.decomposition import PCA
```

```python
pca_X = PCA(n_components=2)
```

```python
pca_X.fit_transform(X_scaled)
```

```python
pca_scaled = pd.DataFrame(data = pca_X.fit_transform(X_scaled),columns=['PCA1','PCA2'])
```

```python
pca_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
sns.scatterplot(pca_scaled.PCA1,pca_scaled.PCA2)
```

```python
wcss=[]
for i in range(1,15):
    km= KMeans(n_clusters=i)
    km.fit(pca_scaled)
    wcss_iter = km.inertia_
    wcss.append(wcss_iter)
```

```python
fig,ax = plt.subplots(figsize = (12, 5))
number_clusters = range(1,15)
plt.plot(number_clusters, wcss, marker = "o")
plt.xlabel("number of clusters")
plt.ylabel('wcss')
```

```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,15)).fit(pca_scaled)
visualizer.show()
```
