# Since we have an identical column, for primary key.
# We will do a merge on the 'Platform column. 
# Default parameter is outer join

```python
# since we have an identical column, for primary key, we will do a merge
pd.merge(console_dates,console_games,left_on ='Platform', right_on = 'Platform', how = 'outer').head()
```

```python
# assigning to a variable 'games' to validate the shape / info of the data set
games = pd.merge(console_dates,console_games,left_on ='Platform', right_on = 'Platform', how = 'outer')
```

# Question 1 : Calculate what % Global Sales were made in North America

```python
games.loc[:,['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum()
```

```python
# summation of the regional sales
sum_sales_region = games.NA_Sales.sum() + games.EU_Sales.sum() + games.JP_Sales.sum() + games.Other_Sales.sum()
```

```python
NA_percet_sales = (games.NA_Sales.sum()/sum_sales_region)*100
```

```python
NA_percet_sales
# 49% of Global sales were made in NA
```

# Question 5 : 
# Order the platforms by thier longeivity innascending order
# is discontinued date - release date

```python
# Question 5 : 
# Order the platforms by thier longeivity innascending order
# is discontinued date - release date
df_Discontinued = pd.to_datetime(games.Discontinued)
df_Release = pd.to_datetime(games.FirstRetailAvailability)
```

```python
# df_Discontinued - df_Release
```

```python
games['longevity'] = df_Discontinued - df_Release
```

```python
games.loc[:,['Discontinued','FirstRetailAvailability','longevity', 'Name']]\
.sort_values(by = 'longevity', ascending = False).head().set_index('Discontinued')
```
